## Variáveis e Tipos de Dados

Variáveis são como caixinhas que guardam informações. Cada caixinha tem um nome e um conteúdo.

```python
x = 10 # Integer (número inteiro) y = 20.5 # Float (número decimal) z = "hello" # String (texto) w = True # Boolean (verdadeiro/falso) print(x) print(y) print(z) print(w) print(type(x)) print(type(y)) print(type(z)) print(type(w))
```

## Operadores Aritméticos

Operadores são símbolos que fazem cálculos, como +, -, *, /

```python
x = 10
y = 5
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x % y)
print(x ** y)
print(x // y)
name = "Peter"
surname = "Gates"
print(name + surname)
print(name * 3)
```

## Questão 1 — Input de dois números e conversão de tipos

```python
x1 = input("Digite o primeiro numero: ")
x2 = input("Digite o segundo numero: ")
print(x1)
print(x2)
print(type(x1))
print(type(x2))
x1 = int(x1)
x2 = int(x2)
print(x1)
print(x2)
print(type(x1))
print(type(x2))
```

## Questão 2 — Comparações e operadores

```python
print(x1 == x2)
print(x1 > x2)
print(x2 > x1)
print(x1 != x2)
x3 = abs(x1 - x2)
print("Diferenca:", x3)
if x1 < x2:
    x1 += x3
else:
    x2 += x3
print("Sao iguais?", x1 == x2)
```
