## Exercício 1 — Comparar dois números

O programa pede dois números e diz qual é o maior, ou se são iguais.

```python
num1 = float(input("Digite o primeiro numero: "))
num2 = float(input("Digite o segundo numero: "))
if num1 > num2: print("O primeiro numero é maior.")
elif num2 > num1: print("O segundo numero é maior.")
else: print("Os dois numeros são iguais.")
```

## Exercício 2 — Classificar nota em letra

O programa pede uma nota numérica e mostra a letra correspondente:
- 90 ou mais: A
- 80 a 89: B
- 70 a 79: C
- 60 a 69: D
- Abaixo de 60: F


```python
percentage = float(input("Digite a nota (0 a 100): "))
if percentage >= 90: grade = "A"
elif percentage >= 80: grade = "B"
elif percentage >= 70: grade = "C"
elif percentage >= 60: grade = "D"
else: grade = "F"
print("Sua nota é:", grade)
```

## Exercício 3 — Vogal ou consoante

O programa pede uma letra e diz se é vogal ou consoante.
Funciona com letras maiúsculas e minúsculas.

```python
letter = input("Digite uma letra: ")
letter = letter.lower()
if letter in "aeiou": print("É uma vogal.")
else: print("É uma consoante.")
```

## Bônus — Par, ímpar ou zero

O programa pede um número inteiro e verifica se é par, ímpar ou zero.
Também verifica se o que foi digitado é realmente um número.

```python
number = input("Digite um numero inteiro: ")

if number.isdigit():
    number = int(number)

    if number == 0:
        print("O numero é zero.")
    elif number % 2 == 0:
        print("O numero é par.")
    else:
        print("O numero é ímpar.")
else:
    print("Isso não é um número inteiro válido.")
```
